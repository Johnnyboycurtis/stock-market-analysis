```python
from query_stocks import load_ticker_symbols, query_stock_prices
```

```python
ticker_symbols = load_ticker_symbols(filename="stock-list.txt")
ticker_symbols.sort()
print(ticker_symbols)
```

```python
price_df = query_stock_prices(tickers=ticker_symbols)
```

```python
price_df.tail()
```

```python
sma_60 = price_df.rolling(50).mean()
sma_60.tail()
```

```python
sma_price_diff = 100*(price_df/ sma_60 - 1).round(3)
```

```python
sma_price_diff.tail(10)
```

```python
last_row = sma_price_diff.iloc[-1]
last_row.name = sma_60.index[-1]
last_row.sort_values(inplace=True)

ax = last_row.plot.bar(figsize=(10, 6))
ax.set_ylim(min(-4, last_row.min() - 2),last_row.max() + 2)
ax.set_title(f"Comparisons Against 60 day Moving Average: {last_row.name.date()}")
ax.set_ylabel("%")
```

```python
ticker_to_plot = 'AMZN' 
```

```python
s1 = price_df[ticker_to_plot]
s2_ma = sma_60[ticker_to_plot].dropna()
s1 = s1.loc[s2_ma.index]

ax = s2_ma.plot(figsize = (10, 6), linewidth=2)
s1.plot(ax=ax, sharex=True, linewidth=2, linestyle="--")

ax.set_title(f"{ticker_to_plot} Price vs 60 Day Moving Average")
ax.set_ylim(s1.min()-5, s1.max()+5)
```

```python
def plot_ma(ticker_to_plot, sma_series, price_df):
    s1 = price_df[ticker_to_plot]
    s2_ma = sma_series[ticker_to_plot].dropna()
    s1 = s1.loc[s2_ma.index]

    ax = s2_ma.plot(figsize = (10, 6), linewidth=2)
    s1.plot(ax=ax, sharex=True, linewidth=2, linestyle="--")

    ax.set_title(f"{ticker_to_plot} Price vs 60 Day Moving Average")
    ax.set_ylim(s1.min()-5, s1.max()+5)
    plt.show()
    return ax
```

```python
ax = plot_ma('AAPL', sma_60, price_df=price_df)
```

```python
# https://plotly.com/python/time-series/

from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)


app.layout = html.Div([
    html.H4('Stock price analysis'),
    dcc.Graph(id="time-series-chart"),
    html.P("Select stock:"),
    dcc.Dropdown(
        id="ticker",
        options=ticker_symbols,
        value="AAPL",
        clearable=False,
    ),
])


@app.callback(
    Output("time-series-chart", "figure"), 
    Input("ticker", "value"))


def display_time_series(ticker):
    df = pd.DataFrame()
    df['MA(50)'] = sma_60[ticker]
    df[ticker] = price_df[ticker]

    df = df.dropna()
    fig = px.line(df)
    return fig


app.run_server(debug=True)
```
